```python
x=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
sq=Sequential()
```

```python
x_train.shape
```

```python
sq.add(Dense(19,activation='relu'))
```

```python
sq.add(Dense(19,activation='relu'))
sq.add(Dense(19,activation='relu'))
sq.add(Dense(19,activation='relu'))
sq.add(Dense(1))
```

```python
sq.compile(loss='mse',optimizer='adam')
```

```python
sq.fit(x_train,y_train,epochs=400,validation_data=(x_test,y_test),batch_size=128)#validation data is validated after each epoch so that
#we can find the overfitted data->val_loss is the loss for test data(both loss are given in sq.history.history )
#batch_size-data are fed in batch wise..smaller the batch size lower is the likelywood to overfit since only small batches are fed
#at once so we can find quickly before overly overfitting
```

```python
his=pd.DataFrame(sq.history.history)
```

```python
pd.DataFrame(sq.history.history).plot()#val_loss is the loss in test data...if it overfits there would be a spike in graph
#technically we can continue training more becoz we didnt overfit yet...but its good enough for now
```

```python
pred=sq.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
mean_absolute_error(y_test,pred)
```

```python
mean_squared_error(y_test,pred)**0.5
```

```python
#prediction is not good enough becoz of using all data which includes out liers
#so use the same prediction with bottom 99 percent
```

```python
explained_variance_score(y_test,pred)#tries to fit 80 percent of the data which includes outliers...which leads to huge error
```
