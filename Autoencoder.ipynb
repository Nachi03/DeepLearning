```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.optimizers import SGD#sarcastic dradient decent to change the learning rate(temp)
#if it starts encoding wrong ,we will reduce the learning rate
```

```python
#seperate encoder
encoder=Sequential()
encoder.add(Dense(2,activation='relu',input_shape=[3]))
```

```python
#seperate decoder
#as we need the autocoder to make sure input == otput so that we can get the Principle component in the hidden layer
decoder=Sequential()
decoder.add(Dense(3,activation='relu',input_shape=[2]))
```

```python
#combine encoder and decoder
autocoder=Sequential([encoder,decoder])
```

```python
#compile on autocoder
autocoder.compile(loss='mse',optimizer=SGD(lr=1.5))
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
scaleddf=sc.fit_transform(df)
```

```python
#fit on autocoder
#as we need the autocoder to make sure input == otput so that we can get the Principle component in the hidden layer
autocoder.fit(scaleddf,scaleddf,epochs=5)
```

```python
encoded=encoder.predict(scaleddf)
#to get the principle components
```

```python
plt.scatter(encoded[:,0],encoded[:,1],c=y)
```
