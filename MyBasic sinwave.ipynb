```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datax=np.linspace(0,50,501)
```

```python
datax
```

```python
data=np.sin(datax)
```

```python
data
```

```python
plt.plot(datax,data)
```

```python
df=pd.DataFrame(data=data,index=datax,columns=['sine'])
```

```python
df
```

```python
test_per=0.1
```

```python
testlen=test_per*(len(df))
```

```python
testlen
```

```python
testlen=np.round(testlen)
```

```python
testlen
```

```python
testlen=int(testlen)
```

```python
train=df.iloc[:len(df)-testlen]#we are using iloc..i.e index
```

```python
test=df.iloc[len(df)-testlen:]
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
#her we trasnform both x and y labels becoz the out (y) is agin fed as input(x) to the mode
```

```python
train=mn.fit_transform(train)
#as usual only fit with train fata to prevent data leakage
```

```python
test=mn.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
#to convert the data into batches that can be used for training
```

```python
length=50
#length is the no of data pt need to get a single trend
batch_size=1
#batch_size-batch of length of data pts hould be passed at a time ..generally 1 is good
```

```python
n_feat=1
```

```python
ts=TimeseriesGenerator(data=train,targets=train,length=length,batch_size=batch_size)
```

```python
len(train)
```

```python
len(ts)
#len(train)-length
#because we dont have 590 (length) val to get the last batch
```

```python
train
```

```python
x,y=ts[0]
#tuple of x and y
```

```python
x
#first 50 val
```

```python
y
#will be the 26 val..and it will be the last input to the next batch(on 25)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import SimpleRNN,Dense
```

```python
sq=Sequential()
```

```python
sq.add(SimpleRNN(length,input_shape=(length,n_feat)))
sq.add(Dense(1,activation='relu'))
sq.compile(loss='mse',optimizer='adam')
```

```python
sq.fit_generator(ts,epochs=5)
```

```python
loss=pd.DataFrame(sq.history.history)
```

```python
loss.plot(figsize=(10,6))
```

```python
current_batch=t
predict=[]

for i in range(length):
    current_predict=sq.predict(current_batch)[0]
    current_batch=np.append(current_batch[:,1:,:],[[current_predict]],axis=1)
    predict.append(current_predict)
```
